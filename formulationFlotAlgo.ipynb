```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=db[['age', 'fare']])
plt.show()
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

db = remove_outliers(db, 'fare')
db = remove_outliers(db, 'age')
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=db[['age', 'fare']])
plt.show()
```

```python
# maintenat la 3eme partie 
from sklearn.preprocessing import StandardScaler, LabelEncoder, KBinsDiscretizer
label_encoder = LabelEncoder()
db['sex'] = label_encoder.fit_transform(db['sex'])
df = pd.get_dummies(db, columns=['embarked'], drop_first=True)
```

```python
# 2. Discrétisation de l'Âge
age_discretizer = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
db['AgeGroup'] = age_discretizer.fit_transform(df[['age']])
```

```python
scaler = StandardScaler()
df['fare'] = scaler.fit_transform(df[['fare']])
```

```python
survival_counts = db['survived'].value_counts()
min_count = survival_counts.min()
df_balanced = db.groupby('survived').apply(lambda x: x.sample(min_count)).reset_index(drop=True)
```
